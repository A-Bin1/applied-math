Perform a Least-Squares regression in Python with an estimation function defined by $\hat{y}(x) = a_1x + a_2$.



```python
#store the x and y values:

import numpy as np

x = np.array([3, 5, 7, 9, 11])
y = np.array([13.2, 17.5, 21.0, 25.9, 29.1])

# assemble matrix A
A = np.vstack([x, np.ones(len(x))]).T

# turn y into a column vector
y = y[:, np.newaxis]
```

```python
from scipy import optimize
import matplotlib.pyplot as plt

plt.style.use( 'seaborn-v0_8-poster')
```

```python
# Direct least square regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
```

Plotting the results of Direct Inverse Method Plot using standard matplotlib.pyplot tools.

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.',label = 'X,Y Coordinates')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.title(f'Plot of Least-Squares Regression: Direct Inverse Method')
plt.xlabel('X axis', fontsize = 13)
plt.ylabel('Y axis', fontsize = 13)
plt.legend(loc = 0)
plt.show()
```
