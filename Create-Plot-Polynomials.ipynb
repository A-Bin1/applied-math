Creating a plot for three distinct polynomials, providing the appropriate axis and labels.

```python
import matplotlib.pyplot as plt
import numpy as np
import math as math
```

```python
def polyn_graph():
    """
    creates a plot for quartic, quadratic within an absolute value,
    and cubic polynomials with 500 sample data points.

    """

    plt.figure(figsize = (10,6))
    x = np.linspace(-5,5,500)
    plt.plot(x, x**4 + 4*x**3 - 3, 'k--', label = 'quartic')
    plt.plot(x,abs(((3*x)**2) - (5*x)+2), 'm-', label = 'absolute value quadratic')
    plt.plot(x, (x**3) - 0.5*x, 'c:', label = ' cubic')
    plt.title(f'Plot of Three Polynomials from {x[0]} to {x[-1]}')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.xlim(-6)
    plt.ylim(-30,25)
    plt.legend(loc = 2)
    plt.show()
```

```python
polyn_graph()
```
