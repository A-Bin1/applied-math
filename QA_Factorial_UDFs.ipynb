Non-Recursive Factorial Function Attempt #1

```python
def factorial_nrec(x):
    """
    takes an integer x and iterates through until reaching 0, creating and appending
    to a list of values and then taking the product of each integer in the list (transformed into an array)
    
    """
    val_lst = []
    while x > 0:
        val_lst.append(x)
        x = x-1
    fact_arry = np.array(val_lst) #converting list of numbers into array
    fact_val = int(np.prod(fact_arry)) #removing the np.64() annotation, converting to integers, calculating factorial
    return(fact_val)
```

```python
factorial_nrec(4)
```

```python
factorial_nrec(20)
```

Recursive Factorial Function from Python Programming and Numerical Methods - A Guide for Engineers and Scientists (Kong, Siauw, Bayen)

```python
def factorial_rec(n):
    """
    Computes and returns the factorial of n, 
    a positive integer.
    
    """
    if n == 1: # Base cases!
        return 1
    else: # Recursive step
        return n * factorial_rec(n - 1) # Recursive call
```

```python
factorial_rec(4) == factorial_nrec(4)
```

```python
factorial_rec(21) == factorial_nrec(21)
```

Is there a better way to check where the function outputs are equal?

```python
for i in range(1,25):
    result = i if factorial_nrec(i) != factorial_rec(i) else None
    print(result)
        
```

Math Module Solution

```python
import math
math.factorial(21) == factorial_rec(21)
```

Updated to list comprehension for efficiency. The recursive function is less lines than both written attempts, but more likely to have performance issues due to the nature of factorials in this case.

```python
result = sum(i for i in range(1,1000) if math.factorial(i) != factorial_rec(i))

print(result)
```

This is a quick way to check if any are wrong, but sums the indices instead of identifying which are incorrect.

Another approach to identify the specific index:

```python
def factorial_check():
    result = sum(i for i in range(1,1000) if math.factorial(i) != factorial_nrec(i))
    index_lst = []
    if result > 0:
        for i in range(1,1000):
            if math.factorial(i) != factorial_nrec(i):
                index_lst.append(i)
    return(index_lst)
```
