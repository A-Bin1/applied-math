Math Model Applications: Budgeting

```python
def travel_cost_calculator(savings, budget, frequency):
    """
    input number for each variable
    savings = given savings at specified frequency (ignoring inflation)
    budget = number to obtain with savings
    frequency = how often are we saving? Index of the frequency_type list
    calculate and rounding appropriately to avoid absurd answers such as 167 months

    """
    frequency_type = ['day', 'week', 'month', 'year']
    frequency_val = round(budget/savings,2)
    if frequency == 0 and frequency_val/7 > 12:
        return(f'It will take about {round(frequency_val/30)} {frequency_type[2]}(s)')
    elif frequency == 1 and frequency_val/4 > 11:
        return(f'It will take about {round(frequency_val/52)} {frequency_type[3]}(s)')
    elif frequency == 2 and frequency_val/12 > 1:
        return(f'It will take about {round(frequency_val/12)} {frequency_type[3]}(s)')
    else:
        return(f'It will take about {round(frequency_val)} {frequency_type[frequency]}(s)')
```

```python
travel_cost_calculator(150, 1500, 2)
```

```python
travel_cost_calculator(680, 25000, 1)
```
