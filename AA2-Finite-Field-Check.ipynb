```python
def check_factors(val):
    primelst = []
    factorlst = []
    # checkpoint = val
    # if val > 1000:
    #     checkpoint = math.trunc(math.sqrt(val)) # removed this section because factors kept being left out. Only needed if doing these by hand.

    for i in range(val):
        if isprime(i):
            primelst.append(i)
    for i in primelst:
        if val % int(i) == 0:
            factorlst.append(i)
    if len(factorlst) == 0:
        return(val)

    

    return(factorlst)

```

```python
#how do we check for the format p^k? One option is to see it from the result of check_factors, but mental math would be needed
#in the case of 100 for example (2^2 x 5^2).

#42.1 Theorem for reference: Let E be a finite extension of degree n over finite field F. If F has q elements, then E has q^n elements.

#Then we are looking for finite field F and we suppose the number given in problems 42.1 - 42.3 are E?

#So then the function must check if only one factor returns --> then it is in the form of q^n elements. 

def finite_field_check(val):
    p = check_factors(val)
    

    if check_factors(val) == val: #distinguishing for this case since check_factors outputs an integer for prime input
        return(f'finite field {p}^1 because {round(p)} is prime.')
    elif type(p) == list and len(p) == 1: #https://www.pythonpool.com/python-division-and-assignment-operator/
        b = p[0]
        n = 0
        new = val
        while new % b == 0:
            new = new /b
            n = n+1
        return(f'finite field {b}^{n}')
    return(f'Not a finite field. Factors are {p}')
```
