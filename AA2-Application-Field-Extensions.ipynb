Find the degree and a basis for the given field extension $Q(∛2,√3)$ over $Q$. Justify your answer.

The minimal polynomial of $∛2$ over $Q$  is $x^3-2 $ since $(∛2)^3-2=0$.

Then, the polynomial has a degree of 3 over $Q$.

Since $√3$ is not in $Q(∛2)$, the degree of the extension is the product of $Q(∛2)$ over $Q$ and $Q(√3)$ over $Q$.

Then, the minimal polynomial of $√3$ over $Q$ is $x^2-3$ since $〖√3〗^2-3=0$ and the degree is 2 over $Q$.

Then, the product of the degrees is $3×2=6$.

Then, by Theorem 40.4, the basis of $Q(∛2,√3)$={$1,∛2,√3,∛2 √3$}.

```python
def min_polynomial_with_root(base, power):
    """
    function takes input of base and value and determines a polynomial where such value is a root
    
    """
    degree = int(power**-1) #because we're focused on square roots, cubed roots, etc. 2^2 would not need to be separated into componenets for this.
    val = base**power
    result = val**degree - base
    if round(val**degree - base,5) == 0: #trying to avoid rounding differences
        return(f"({base}^({Fraction(power).limit_denominator(1_000_000)}))^{degree} - {base}")
    else:
        return(result)
    
```

```python
def min_polynomial(base, power):
    """
    function takes input of base and value and determines a polynomial where such value is a root
    
    """
    degree = power**-1 #because we're focused on square roots, cubed roots, etc. 2^2 would not need to be separated into componenets for this.
    val = base**power
    result = val**degree - base
    if round(val**degree - base,5) == 0: #trying to avoid rounding differences
        return(f"x^{int(degree)} - {base}")
    else:
        return(result)
```

```python
#combining these components altogether into one output

def basis_and_degreeQ(val1_base, val1_power, val2_base, val2_power):
    degree1 = int(val1_power**-1)
    degree2 = int(val2_power**-1)
    val1 = val1_base**val1_power
    val2 = val2_base**val2_power
    product_degree = int(degree1*degree2)
    written_form_val1 = f'{val1_base}^({Fraction(val1_power).limit_denominator(1_000_000)})'
    written_form_val2 = f'{val2_base}^({Fraction(val2_power).limit_denominator(1_000_000)})'
    nl = '\n'

    if type(min_polynomial_with_root(val1_base, val1_power)) == str and type(min_polynomial_with_root(val2_base, val2_power) == str):
        proof_poly1 = print(f'The minimal polynomial of {written_form_val1} over Q is {min_polynomial(val1_base, val1_power)} ' 
               f'since {min_polynomial_with_root(val1_base, val1_power)} = 0.{nl}'
               f'Then, the polynomial has a degree {degree1} over Q.{nl}')
    if val_check(val1,val2) == False:
        proof_poly2 = print(f'Since {written_form_val2} is not in Q({written_form_val1}), '
                            f'the degree of the extension is the product of Q({written_form_val1}) over Q '
                            f'and Q({written_form_val2}) over Q.{nl}'
                            f'Then, the minimal polynomial of {written_form_val2} over Q is {min_polynomial(val2_base, val2_power)} '
                            f'since {min_polynomial_with_root(val2_base, val2_power)} = 0.{nl}'
                            f'Then, the polynomial has a degree {degree2} over Q.{nl}'
                            f'Then, the product of the degrees is {degree1} x {degree2} = {product_degree}.{nl}')
    
    #the basis includes the multiplicative inverse of Q which is 1. Then the two given values along with the product of both values.
    basis = f"{{1,{written_form_val1},{written_form_val2}, ({written_form_val1})({written_form_val2})}}"

    proof_basis = f'Then, by Theorem 40.4, the basis Q({written_form_val1},{written_form_val2}) = {basis}'

    print(proof_basis)

```
