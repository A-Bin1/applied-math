```python
#put it altogether

def create_matrix(num_range):
    """
    user inputs number range for x,y,z values
    create list of size generated as a random number, converted into x, y, z 
    and transformed into array.
    """
    multiplier = np.random.randint(15,25)
    multiplier = multiplier *3
    xyz_lst = np.random.randint(0,num_range,multiplier) #make this into sets of 3
    n = 3

#splitting list into xzy chunks https://www.geeksforgeeks.org/python/how-to-split-lists-in-python/
# Number of Sets
# Splitting Sets and Storing them in List
    xyz_sys = [xyz_lst.tolist()[i:i + n] for i in range(0, len(xyz_lst), n)]
    matrix = []
    n_lst = int(len(xyz_sys))
    for i in range(n_lst):
        matrix.append(xyz_sys[i])
    
    matrix = np.array(matrix)
    return(matrix)
```
