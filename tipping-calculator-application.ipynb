User Input, Branching, List Comprehension Review Application: Tipping Calculator

```python
def tippingCalc():
    """
    user provides service experience input given an assumed list of negative words to determine
    what the tip should be

    it is assumed if the service was up to par or exceeded expectations, a base 25% tip is applied

    user provides input of total restaurant bill
    
    """
    service_input = input("Enter your experience: ")
    price_input = input("Enter your restaurant bill: ")
    service_lst = [service_input]
    price = price_input
    service_bad = ["late",  "bad", "slow"]
    service_awful = ["wrong", "cold","horrible", "rude"]
    tip = float(price)*0.25
    if any(item in service_lst for item in service_bad):
        tip  = tip*0.85
    elif any(item in service_lst for item in service_awful):
        tip = tip*0.2
    else:
        tip = tip*.1
    return(f"The tip should be ${round(tip, 2)} based on the given feedback.")
```

```python
tippingCalc()
```
