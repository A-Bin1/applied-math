Application 1: Designated Bounds

Linear expression to maximize profit: 20x + 30y

Material constraint: 20x + 30y <= 690

Labor constraint: 5x + 4y <= 120

Product 1 bound: x >= 4

Product 2 bound: y >= 2

```python
from scipy.optimize import linprog
```

```python
obj = [-20,-30]
```

```python
# Constraints Left side x and y Coefficient matrix
lhs_ineq = [[20,  30], [5,  4]]  
```

```python
rhs_ineq = [[690, 120]]
```

```python
xy_bounds = [(4,None), (None,2)] #putting 0 yields an infeasible result
```

```python
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,bounds = xy_bounds ,method="highs")
```

```python
print(opt)
```

Experimenting with Multiple of Inputs for X,Y bounds.

Example: Product 1 now has a contract of at least 3 per week and Product 2 has a contract for at least 1 a week.

Linear expression to maximize profit: 20x + 30y

Material constraint: 20x + 30y <= 690

Labor constraint: 5x + 4y <= 120

Product 1 bound: x >= 3

Product 2 bound: y >= 1

This could be repeated as the prior problems, but storing different scenarios of these bounds to iterate through is more efficient.

```python
obj = [-20,-30]
lhs_ineq = [[20,  30], [5,  4]]
rhs_ineq = [[690, 120]]
xy_bounds = [(4,None), (None,2)]
```

Refresher on enumerate for proper indexing:

```python

xy_arr = [[(4,None), (None,2)],[(3, None),(None, 1)],[(5, None),(None, 4)]]

def lin_prog_max_sim(arr):
    """
    input contains constant constraints with a list of bounds
    using arr to keep bound, list, list of lists separate, easier verbiage (list of arrays was the goal)
    iterate through each set of bounds and append corresponding values
    create xyz list of lists

    output is resulting maximization

    """
    xyz_val = []

    for i, v in enumerate(arr):
        out = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,bounds = arr[i] ,method="highs")
        xy_arr = out.x
        z_arr = out.fun
        x_val = float(xy_arr[0]) #need to capture an index of the results for each variable
        y_val = float(xy_arr[1])
        z_val = float(z_arr)
        xyz_val.append(x_val)
        xyz_val.append(y_val)
        xyz_val.append(z_val)
   
    #splitting list into xzy chunks https://www.geeksforgeeks.org/python/how-to-split-lists-in-python/
    # Number of Chunks
    n = 3

    # Splitting Chunks and Storing them in List
    xyz_val_chunks = [xyz_val[i:i + n] for i in range(0, len(xyz_val), n)]



    return(xyz_val_chunks)


```

```python
lin_prog_max_sim(xy_arr)
```
